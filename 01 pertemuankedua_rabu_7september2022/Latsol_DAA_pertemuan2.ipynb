```python
def getBagi(myList):
  sum = 0
  for row in myList:
    for item in row:
      sum /= item
  return sum

getBagi([[5],[9]])
```

```python
def getKurang(myList):
  sum = 0
  for row in myList:
    for item in row:
      sum -= item
  return sum
getKurang([[45,11],[3,7]])
```
